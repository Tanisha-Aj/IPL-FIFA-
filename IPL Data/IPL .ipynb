```python
#number of times a particular player got man of match OR frequency of most man of match awards
ipl['player_of_match'].value_counts()
```

```python
#now i want to make bar plot for this top 5 man of match
list(ipl['player_of_match'].value_counts()[0:5].keys()) #this gives categorical value for 1st parameter of bar plt
```

```python
plt.figure(figsize=(9,6))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color="b")
plt.show()        
```

```python
#finding out the number of toss WINS w.r.t each team
ipl['toss_winner'].value_counts()
```

```python
#extracting the records where a team won batting first
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head(10)
```

```python
#making a HISTOGRAM
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs'])
plt.title("distributon of Runs")
plt.xlabel("runs")
plt.show
```

```python
#histogram shows that about 100 matches where team won the run by some 0 to 20 runs only
#2 or 3 matches where team won by 140 runs 
```

```python
(list(batting_first['winner'].value_counts()[0:5].keys()))
```

```python
batting_first['winner'].value_counts()
```

```python
(list(batting_first['winner'].value_counts()[0:5]))
```

```python
#BAR PLOT for top 3 teams with most wins after batting first
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["blue","yellow","orange"])
plt.show()
```

```python
#Making a pie chart
plt.figure(figsize=(6,6))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
